```python
# check installed version (must be >3.0)
import pycaret
pycaret.__version__
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
parkinsons_telemonitoring = fetch_ucirepo(id=189)

# data (as pandas dataframes)
X = parkinsons_telemonitoring.data.features
y = parkinsons_telemonitoring.data.targets

# Print metadata
print("Metadata:")
print(parkinsons_telemonitoring.metadata)

# Print variable information
print("\nVariable Information:")
print(parkinsons_telemonitoring.variables)
```

```python
# exclude one traget (motor_UPDRS)
y = parkinsons_telemonitoring.data.targets['total_UPDRS']

```

```python
from pycaret.regression import setup, compare_models

# Initialize the setup
regression_setup = setup(data=X, target=y)

# Compare models to find the best one
best_model = compare_models()
```

```python
# Here we create and tune a specific model
## Here we see the Decision Tree Regressor makes more accurate perdictions with a MAE of 0.9316

## We choose the Decision Tree Regressor
from pycaret.regression import create_model, tune_model

# Create a specific model, e.g., Decision Tree Regressor
model = create_model('dt')

# Tune the model for better performance
tuned_model = tune_model(model)

```

```python
## Here we will evaluate the model
from pycaret.regression import evaluate_model
from pycaret.regression import save_model
from pycaret.regression import finalize_model

evaluate_model(tuned_model)
final_model = finalize_model(tuned_model)

## Now we save the model.
# Save the model
save_model(final_model, 'parkinsons_model')
```
